# About Data

Kaggle competition - https://www.kaggle.com/competitions/kkbox-churn-prediction-challenge

KKBOX is Asia’s leading music streaming service, holding the world’s most comprehensive Asia-Pop music library with over 30 million tracks. They offer a generous, unlimited version of their service to millions of people, supported by advertising and paid subscriptions. This delicate model is dependent on accurately predicting churn of their paid users.


In this competition you’re tasked to build an algorithm that predicts whether a user will churn after their subscription expires. Currently, the company uses survival analysis techniques to determine the residual membership life time for each subscriber. By adopting different methods, KKBOX anticipates they’ll discover new insights to why users leave so they can be proactive in keeping users dancing.

**Churn Definition:**
    ```The criteria of "churn" is no new valid service subscription within 30 days after the current membership expires.```

# Data Visualization

````python
churn_counts = df['is_churn'].value_counts()
gender_counts = df['gender'].fillna('Missing').value_counts()

# Plotting the pie charts
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

axes[0].pie(churn_counts, labels=churn_counts.index.map({0: 'Not Churn', 1: 'Churn'}), autopct='%1.1f%%', startangle=90, colors=['#8fd9b6','#ffcc99'])
axes[0].set_title('Churn/Not Churn')

axes[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#d3d3d3', '#c2c2f0','#ffb3e6',])
axes[1].set_title('Gender Distribution')

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(12, 8))

reg_years_in_data = df.groupby("registration_year")["msno"].count().reset_index()
sns.barplot(x="registration_year", y="msno", data=reg_years_in_data)
plt.title('Number of Registrations per Year')
plt.xlabel('Registration Year', fontsize=14)
plt.ylabel('Count of Registrations', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)# Rotate x-axis labels if necessary
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
churn_counts = df.groupby("registration_year")["is_churn"].sum().reset_index()
churn_counts.columns = ["registration_year", "churn_count"]

# Mean churn rate per year
mean_churn_rate = df.groupby("registration_year")["is_churn"].mean().reset_index()
mean_churn_rate.columns = ["registration_year", "mean_churn_rate"]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=False)

# First plot: Number of churns per year
sns.barplot(ax=axes[0], x="registration_year", y="churn_count", data=churn_counts)
axes[0].set_title('Number of Churns per Year', fontsize=16)
axes[0].set_xlabel('Registration Year', fontsize=14)
axes[0].set_ylabel('Count of Churns', fontsize=14)
axes[0].tick_params(axis='x', rotation=45, labelsize=12)
axes[0].tick_params(axis='y', labelsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Second plot: Mean churn rate per year
sns.barplot(ax=axes[1], x="registration_year", y="mean_churn_rate", data=mean_churn_rate, color='grey')
axes[1].set_title('Mean Churn Rate per Year', fontsize=16)
axes[1].set_xlabel('Registration Year', fontsize=14)
axes[1].set_ylabel('Mean Churn Rate', fontsize=14)
axes[1].tick_params(axis='x', rotation=45, labelsize=12)
axes[1].tick_params(axis='y', labelsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
````

### Last Transaction date & Life Time of users

````python
df['actual_amount_paid_sum'].mean()
````

````python
churned_df = df[df["is_churn"] == 1]

life_time_of_churned_users = churned_df["usage_period_days"]

plt.figure(figsize=(12, 6))
sns.histplot(life_time_of_churned_users, bins=60, kde=True)
plt.title('Distribution of Lifetime of Churned Users', fontsize=16)
plt.xlabel('Lifetime (Days)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(np.linspace(0, 5000, 21), rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
ltv_for_users = churned_df["actual_amount_paid_sum"] * (1 - churned_df["is_cancel_mean"])
churned_df["LTV"] = churned_df["actual_amount_paid_sum"] * (1 - churned_df["is_cancel_mean"])
ltv_for_users_filt = ltv_for_users[ltv_for_users < ltv_for_users.quantile(0.99)]

plt.figure(figsize=(10, 6))
sns.histplot(ltv_for_users_filt, bins=20, color='darkgreen', kde=True)
plt.title('Distribution of Lifetime Value (LTV) for Users', fontsize=16)
plt.xlabel('LTV', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

````

````python
ltv_for_users.quantile([0.2, 0.5, 0.6, 0.7, 0.9, 0.95, 0.99, 0.995])
````

````python
ltv_by_cities = churned_df.groupby("city")["LTV"].mean().reset_index()
cities_counts = df.groupby("city")["msno"].count().reset_index()
````

````python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 11))

# Plot 1: Cities of users
sns.barplot(ax=axes[0], x="city", y="msno", data=cities_counts, palette="magma")
axes[0].set_title('Cities of users', fontsize=16)
axes[0].set_xlabel('City encoded', fontsize=14)
axes[0].set_ylabel('Count of Registrations', fontsize=14)
axes[0].tick_params(axis='x', rotation=45, labelsize=12)
axes[0].tick_params(axis='y', labelsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Plot 2: LTV by cities
sns.barplot(ax=axes[1], x="city", y="LTV", data=ltv_by_cities, color="gold")
axes[1].set_title('LTV by cities', fontsize=16)
axes[1].set_xlabel('City encoded', fontsize=14)
axes[1].set_ylabel('Average Life-Time Value (LTV)', fontsize=14)
axes[1].tick_params(axis='x', rotation=45, labelsize=12)
axes[1].tick_params(axis='y', labelsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
````

## Correlation Analysis

````python

numeric_df = df.select_dtypes(include=[np.number])

numeric_df["curr_LTV"] = numeric_df["actual_amount_paid_sum"] * numeric_df["is_cancel_mean"]


churn_correlations = numeric_df.corr()["is_churn"]
ltv_correlations = numeric_df.corr()["curr_LTV"]

filtered_churn_correlations = churn_correlations[churn_correlations.abs() > 0.1].sort_values()
filtered_ltv_correlations = ltv_correlations[ltv_correlations.abs() > 0.1].sort_values()

fig, axes = plt.subplots(1, 2, figsize=(18, 8))  # 1 row, 2 columns

sns.barplot(x=filtered_churn_correlations.values, y=filtered_churn_correlations.index, palette='coolwarm', ax=axes[0])
axes[0].set_title('Significant Correlations with is_churn (>0.1)', fontsize=14)
axes[0].set_xlabel('Correlation Coefficient')
axes[0].set_ylabel('Features')

sns.barplot(x=filtered_ltv_correlations.values, y=filtered_ltv_correlations.index, palette='inferno', ax=axes[1])
axes[1].set_title('Significant Correlations with curr_LTV (>0.1)', fontsize=14)
axes[1].set_xlabel('Correlation Coefficient')
axes[1].set_ylabel('Features')

# Adjusting layout
plt.tight_layout()
plt.show()
````

## Discussion and findings

We have done a basic exploration of our data and patterns that affect *churn rate*. Here's the list of findings and insights that
might be useful for management and for futher model development:

 **Findings**
 1) On average,  churn rate is about **9%** per month. As far as i undersand, it means that on average, about 9% of active users do not renew subscription for next period.
 2) We have many users that **do not specify their *gender* (54%)**
 3) From the data, we easilly see that  **KKBOX service actually keep growing and developing** since more and more users come. It looks
    like more users tend to come (our plot is build on data including only beggining of 2017)

 4) The highest average churn is detected in **2009, 2010 and 2017**.
 5) Distribution of **Usage period is somewhat expected**. Many people who used service for short period of time (less than year or year).
    And smaller amount of people who stayed with service much longer (therefore paid more money).
 6) **Life-time Value** of users is also expected. Many users didn't use service for long time and didn't pay much money.
    However, we have few users that stay with us for much longer and brought more money. Such users are loyal and important for
    us.

 8) Most of the people come from **City 1**. At the same time, average **LTV** for this city is **fairly low**. For some reason,
    people from that city tend to either use service for short period of time or pay less for subscription.
